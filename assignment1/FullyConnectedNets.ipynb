As a sanity check, run the following to check the initial loss and to gradient check the network both with and without regularization. This is a good way to see if the initial losses seem reasonable.

For gradient checking, you should expect to see errors around 1e-7 or less.

````python
np.random.seed(231)
N, D, H1, H2, C = 2, 15, 20, 30, 10
X = np.random.randn(N, D)
y = np.random.randint(C, size=(N,))

for reg in [0, 3.14]:
    print("Running check with reg = ", reg)
    model = FullyConnectedNet(
        [H1, H2],
        input_dim=D,
        num_classes=C,
        reg=reg,
        weight_scale=5e-2,
        dtype=np.float64
    )

    loss, grads = model.loss(X, y)
    print("Initial loss: ", loss)

    # Most of the errors should be on the order of e-7 or smaller.
    # NOTE: It is fine however to see an error for W2 on the order of e-5
    # for the check when reg = 0.0
    for name in sorted(grads):
        f = lambda _: model.loss(X, y)[0]
        grad_num = eval_numerical_gradient(f, model.params[name], verbose=False, h=1e-5)
        print(f"{name} relative error: {rel_error(grad_num, grads[name])}")
````

Did you notice anything about the comparative difficulty of training the three-layer network vs. training the five-layer network? In particular, based on your experience, which network seemed more sensitive to the initialization scale? Why do you think that is the case?

## Answer:

Yes, I noticed that training the five-layer network was more challenging and the network was more sensitive to the initialization scale.
In the three-layer network, it was easier to find values for learning rate and weight scale that led to fast convergence, but in contrast, in a five-layer network, an inappropriate choice of one of the values led to training not progressing, the loss decreasing slowly, or the process appearing unstable.

The reason - the deeper the network, the more layers the gradients pass through. every small error in initialization accumulates throughout the network. If we initialize the weights too low, the gradients tend to fade (vanishing gradients), and if we initialize them too high, the values can explode ( exploding gradients), so the choice of values is  critical when the network is deeper.

## Inline Question 2:

AdaGrad, like Adam, is a per-parameter optimization method that uses the following update rule:

```
cache += dw**2
w += - learning_rate * dw / (np.sqrt(cache) + eps)
```

John notices that when he was training a network with AdaGrad that the updates became very small, and that his network was learning slowly. Using your knowledge of the AdaGrad update rule, why do you think the updates would become very small? Would Adam have the same issue?


## Answer:

In AdaGrad, cache is the sum of the squares of the gradients throughout training. because cache only grows over time (with no decay) the variable in the denominator :  (np.sqrt(cache) + eps), just get larger. As a result, the size of the weight update get smaller even if the gradient is still large. After many iterations the effective learning rate become small and the training become slowly or even stop.

In Adam, this problem almost does not exist because Adam use a moving average (with decay) of the squares of the gradients, rather than sum like AdaGrad. In addition, Adam has a bias correction at the beginning of training so the values in the denominator grow up to a certain limit therefore the updates remain on a reasonable scale over time. As a result, Adam has more stable learning rate and its better suited for training deep networks over time.

````python
from icv83551.classifiers.fc_net import FullyConnectedNet

best_model = None

################################################################################
# TODO: Train the best FullyConnectedNet that you can on CIFAR-10. You might   #
# find batch/layer normalization and dropout useful. Store your best model in  #
# the best_model variable.                                                     #
################################################################################
# *****START OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****

best_val = -1
num_train = 500

# like the three-layer and five-layer we did before,
# we take small subset from the data and train on it to find the best parameters
# for the full training data
mini_data = {
  'X_train': data['X_train'][:num_train],
  'y_train': data['y_train'][:num_train],
  'X_val': data['X_val'],
  'y_val': data['y_val'],
}
# number of random parameters trials
num_trials = 50

for trial in range(1, num_trials+1):
  # randomly parameter in certain range in each trial
  lr = 10 ** np.random.uniform(-4, -2.5)
  ws = 10 ** np.random.uniform(-2.3, -1)
  reg = 10 ** np.random.uniform(-5, 2)
  # no need to use dropout in this stage (as stated up here)
  kr = 1

  # define net with 4 hidden layers
  model = FullyConnectedNet(
    [256, 128, 64, 64],
    weight_scale=ws,
    reg=reg,
    dropout_keep_ratio=kr,
    normalization=None,
    dtype=np.float64
  )

  # use adam optimizer after we set it properly
  solver = Solver(
    model,
    mini_data,
    num_epochs=10,
    batch_size=200,
    update_rule='adam',
    optim_config={'learning_rate': lr},
    verbose=False
  )

  solver.train()

  # compute the best val_acc and save the best parameters
  val_acc = solver.best_val_acc
  print(f'Trial={trial} lr={lr:.1e}, ws={ws:.1e}, reg={reg:.1e}, val_acc={val_acc:.6f}')
  if val_acc > best_val:
    best_val = val_acc
    best_model = model
    best_lr = lr
    best_ws = ws
    best_reg = reg

print(f'\nBest validation accuracy on mini_data: {best_val}')
print(f'Best parameters: lr={best_lr:.1e}, ws={best_ws:.1e}, reg={best_reg:.1e}')

# train model on full data using the best parameters
print('\nTrain full data with best parameters:')
best_model = FullyConnectedNet(
    [256, 128, 64, 64],
    weight_scale=best_ws,
    reg=best_reg,
    dropout_keep_ratio=kr,
    normalization=None,
    dtype=np.float64
  )

best_solver = Solver(
  best_model,
  data,
  num_epochs=10,
  batch_size=200,
  update_rule='adam',
  optim_config={'learning_rate': best_lr},
  verbose=False
)

best_solver.train()

best_val_acc = best_solver.best_val_acc
print(f'\nBest validation accuracy on full data: {best_val_acc}')
print(f'Best parameters: lr={best_lr:.1e}, ws={best_ws:.1e}, reg={best_reg:.1e}')

# *****END OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****
################################################################################
#                              END OF YOUR CODE                                #
################################################################################
````
